```python
def y_to_pose (y):
    if(y==0): pose = 'Plank_Pose_or_Kumbhakasana_'    
    if(y==1): pose = 'Warrior_I_Pose_or_Virabhadrasana_I_'  
    if(y==2): pose = 'Standing_big_toe_hold_pose_or_Utthita_Padangusthasana'     
    if(y==3): pose = 'Warrior_II_Pose_or_Virabhadrasana_II_'     
    if(y==4): pose = 'Cat_Cow_Pose_or_Marjaryasana_'
    if(y==5): pose = 'Child_Pose_or_Balasana_'  
    if(y==6): pose = 'Cobra_Pose_or_Bhujangasana_'     
    if(y==7): pose = 'Tree_Pose_or_Vrksasana_'     
    if(y==8): pose = 'Downward-Facing_Dog_pose_or_Adho_Mukha_Svanasana_'
    return pose
```

```python
import os
import matplotlib.pyplot as plt
import cv2
import mediapipe as mp

mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5, model_complexity=2)

for idx, row in wrong_predicted.iterrows():
    file_name = row['file_name']
    pose_name = str(y_to_pose(row['y_test']))
    for root, dirs, files in os.walk(folder_path):
        if pose_name in dirs:
            subfolder_path = os.path.join(root, pose_name)
            for name in os.listdir(subfolder_path):
                if file_name == name:
                    img_path = os.path.join(subfolder_path, name)
                    img = cv2.imread(img_path)
                    
                    results = pose.process(img)
                    mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
                    
                    plt.axis('off')
                    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                    plt.title(f"File name: {file_name}\nTrue label: {pose_name}\nPredicted label: {y_to_pose(row['y_pred'])}")
                    plt.show()
```
