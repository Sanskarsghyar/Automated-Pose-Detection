```python
from PIL import Image
from tqdm.notebook import tqdm

for i in tqdm(range(len(file_list))):
    f = open("/content/drive/MyDrive/UGP23/dataset/Yoga-82/yoga_dataset_links/" + file_list[i], 'r');
    lines = f.readlines();
    f.close();
    # print(f)
    
    for j in tqdm(range(len(lines))):
        #print(lines[j])
        # print(len(lines[j]))
        img_path, link = lines[j].strip().split("\t")
        # print(img_path)
        # print(link)
        
        folder_name, img_name = img_path.split("/");
        if(j == 0):
            if(not os.path.isdir("/content/drive/MyDrive/UGP23/dataset/Yoga-82/Images/" + folder_name)):
                os.mkdir("/content/drive/MyDrive/UGP23/dataset/Yoga-82/Images/" + folder_name)
        # print("wget -O " + "/content/drive/MyDrive/UGP23/dataset/Yoga-82/Images/" + folder_name + "/" + img_name + " " + link)
        os.system("wget -O " + "/content/drive/MyDrive/UGP23/dataset/Yoga-82/Images/" + folder_name + "/" + img_name + " " + link)
        try:
            img = Image.open("Images/" + folder_name + "/" + img_name)
            img.verify()     # to veify if its an img
            img.close()     #to close img and free memory space
        except (IOError, SyntaxError) as e:
            pass
```
